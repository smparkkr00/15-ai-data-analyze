```python
# 시리즈

## 외부라이브러리 pandas 설치
## pip install pandas

## 외부 라이브러리 불러오기

import pandas as pd

dict_data ={
    "a": 1,
    "b": 2,
    "c": 3
}

sr = pd.Series(dict_data)

print(type(sr))

print(sr)
```

```python
# 원소 선택

import pandas as pd

tup_data = ('영인', '2010-05-01', '여', True)

sr = pd.Series(tup_data, index = ['이름', '생년월일', '성별', '학생여부'])
print(sr)
```

```python
#원소 선택

print(sr.iloc[0])
print(sr['이름'])
```

```python
print(sr.iloc[[1,2]])
print("\n")
print(sr[["생년월일", "성별"]])
```

```python
print(sr)
print("\n")
print(sr[1:2])
print("\n")
print(sr["생년월일":"성별"])
```
