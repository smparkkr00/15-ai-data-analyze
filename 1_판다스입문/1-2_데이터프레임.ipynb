```python
# 데이터프레임
## 딕셔너리 -> 데이터프레임으로 변환
## pandas.DataFrame(딕셔너리)

import pandas as pd

dict_data = {
    "c0": [1, 2, 3], 
    "c1": [4, 5, 6], 
    "c2": [7, 8, 9], 
    "c3": [10, 11, 12], 
    "c4": [13, 14, 15]
}

## 시리즈로 변환한 값
sr = pd.Series(dict_data)
print(sr)

## 데이터프레임으로 변환한 값
df = pd.DataFrame(dict_data)
print(df)
```

```python
import pandas as pd 

exam_data = {
    "이름": ["서준", "우현", "인아"],
    "수학": [90, 80, 70], 
    "영어": [98, 89, 95], 
    "음악": [95, 95, 100], 
    "체육": [100, 90, 90]
}

df = pd.DataFrame(exam_data)
df = df.set_index("이름")
df
```

```python
## 인덱스이름: DF객체.loc[행인덱스, 열이름]
## 정수위치인덱스: DF객체.iloc[행번호, 열번호]

print(df.loc["서준", "음악"])
print(df.iloc[0, 2])
```
