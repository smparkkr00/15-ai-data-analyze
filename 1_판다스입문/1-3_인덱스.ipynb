```python
# 행 인덱스 재배열

import pandas as pd

dict_data = {
    "c0": [1, 2, 3],
    'c1': [4, 5, 6],
    'c2': [7, 8, 9],
    'c3': [10, 11, 12],
    'c4': [13, 14, 15]
}

df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])
df
```

```python
new_index = ['r0', 'r1', 'r2', 'r3', 'r4']
ndf = df.reindex(new_index)
ndf
```

```python
new_index = ['r0', 'r1', 'r2', 'r3', 'r4']
ndf2 = df.reindex(new_index, fill_value='')
ndf2
```

```python
# 행 인덱스 초기화
## df객체.reset_index()

import pandas as pd

dict_data = {
    "c0": [1, 2, 3],
    'c1': [4, 5, 6],
    'c2': [7, 8, 9],
    'c3': [10, 11, 12],
    'c4': [13, 14, 15]
}

df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])

print(df)
```

```python
ndf = df.reset_index()
print(df)
print(ndf)
```

```python
ndf2 = df.reset_index(names = ['c00'])
print(ndf2)
```

```python
ndf3 = df.reset_index(drop = True)
ndf3
```

```python
# 행 인덱스를 기준으로 데이터프레임 정렬
## df객체.sort_index()

import pandas as pd

dict_data = {
    "c0": [1, 2, 3],
    'c1': [4, 5, 6],
    'c2': [7, 8, 9],
    'c3': [10, 11, 12],
    'c4': [13, 14, 15]
}

df = pd.DataFrame(dict_data, index=['r0', 'r2', 'r1'])
df
```

```python
print(df.sort_index())
print(df)
ndf = df.sort_index()
ndf
```

```python
ndf2 = df.sort_index(ascending=False)
ndf2
```

```python
# 열 기준 정렬
## df객체.sort_values()

import pandas as pd

dict_data = {
    "c0": [1, 2, 3],
    'c1': [4, 5, 6],
    'c2': [7, 8, 9],
    'c3': [10, 11, 12],
    'c4': [13, 14, 15]
}

df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])
df
```

```python
ndf = df.sort_values("c2")
ndf
```

```python
ndf2 = df.sort_values("c2", ascending = False)
ndf2
```

```python
ndf4 = df.sort_values(by = ['c3', 'c4'], ascending=[False, True])
ndf4
```
