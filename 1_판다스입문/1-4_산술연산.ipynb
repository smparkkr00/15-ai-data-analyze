```python
# 시리즈 vs 시리즈

import pandas as pd

dict_data1 = {
    '국어': 100,
    '영어': 80,
    '수학': 90
}

dict_data2 = {
    '수학': 80,
    '국어': 90,
    '영어': 80
}

student1 = pd.Series(dict_data1)
student2 = pd.Series(dict_data2)

print(student1)
print()
print(student2)
```

```python
addition = student1 + student2

print(type(addition))
print(addition, '\n')

substraction = student1 -student2
multiplication = student1 * student2
division = student1/ student2

print(substraction, '\n')
print(multiplication, '\n')
print(division, '\n')
```

```python
lis = [addition, substraction, multiplication, division]

result = pd.DataFrame(lis, index = ['덧셈', '뺄셈', '곱셈', '나눗셈'])

result
```

```python
# NaN 값이 있는 시리즈 연산

import pandas as pd
import numpy as np

dict_data1 = {
    '국어': np.nan,
    '영어': 80,
    '수학': 90
}

dict_data2 = {
    '수학': 80,
    '국어': 90
}


student1 = pd.Series(dict_data1)
student2 = pd.Series(dict_data2)

print(student1)
print(student2)
```

```python
addition = student1 + student2
substraction = student1 - student2
multiplication = student1 * student2
division = student1 / student2

print(addition, '\n')
print(substraction, '\n')
print(multiplication, '\n')
print(division, '\n')
```

```python
result = pd.DataFrame([addition, substraction, multiplication, division], ['덧셈', '뺄셈', '곱셈', '나눗셈'])
result
```

```python
# 연산 메서드

import pandas as pd
import numpy as np

dict_data1 = {
    '국어': np.nan,
    '영어': 80,
    '수학': 90
}

dict_data2 = {
    '수학': 80,
    '국어': 90
}


student1 = pd.Series(dict_data1)
student2 = pd.Series(dict_data2)

sr_add = student1.add(student2, fill_value=0)
sr_sub = student1.sub(student2, fill_value=0)
sr_mul = student1.mul(student2, fill_value=0)
sr_div = student1.div(student2, fill_value=0)

lis = [sr_add, sr_sub, sr_mul, sr_div]

result = pd.DataFrame(lis, ['덧셈', '뺄셈', '곱셈', '나눗셈'])
result
```

```python
# 데이터프레임 vs 데이터프레임

import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")
df = titanic.loc[ : ,['age', 'fare']]

df.tail()
```

```python
addition = df + 10

addition.tail()
```

```python
substraction = addition - df
print(type(substraction))
substraction.tail()
```

```python
sample1 = addition.tail(10)
sample1.shape
sample1
```

```python
sample2 = substraction.tail(10).fillna(0)
sample2
```

```python
df_add = sample1.add(sample2, fill_value=0)
df_sub = sample1.sub(sample2, fill_value=0)
df_mul = sample1.mul(sample2, fill_value=0)
df_div = sample1.div(sample2, fill_value=0)

print(df_add)
print()
print(df_sub)
print()
print(df_mul)
print()
print(df_div)
print()
```
