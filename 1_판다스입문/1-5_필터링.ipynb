```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")

# 행 인덱스 0~9 범위에서 age, fare 2개열을 선택하여 데이터프레임 만들기

df = titanic.loc[0:9, ['age', 'fare']]

df
```

```python
# 불린 인덱싱(boolean indexing)

# 조건식 적용 (20세 미만, age < 20)

df['age'] < 20

# 불린 인덱싱(20세 미만의 조건을 충족하는 행을 필터링)

cond = df['age'] < 20
df[cond]
```

```python
df.loc[cond]
```

```python
maks1 = (titanic.age >= 10) & (titanic.age < 20)
df_teenage = titanic[maks1]
df_teenage.head()
```

```python
# 나이가 10세미만이고 여성인 승객

mask2 = (titanic.age < 10) & (titanic.sex == 'female')
df_female_under10 = titanic.loc[mask2]
df_female_under10.head()
```

```python
mask3 =  (titanic.age < 10) |  (titanic.age >= 60)
df_under10_morethan60 = titanic.loc[mask3,['age', 'sex', 'alone']]
df_under10_morethan60
```

```python
# query() 메서드 활용

import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")

# 행 인덱스 0~9 범위에서 age, fare 2개열을 선택하여 데이터프레임 만들기

df = titanic.loc[ : , ['age', 'fare']]

df
```

```python
# 20세 미만의 조건을 충족하는 행을 필터링

df.query('age < 20')
```

```python
# 20세 이상이고 요금이 15미만인 조건을 모두 충족하는 행을 필터링

df.query("age >= 20 and fare < 15")
```

```python
df.query("age >= 20 or fare < 15")
```

```python
# isin() 메서드 활용

## df객체.isin(추출 값의 리스트)

import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")

pd.set_option("display.max_columns", 10)

mask1 = titanic['embark_town'] == 'Southampton'
mask2 = titanic['embark_town'] == 'Queenstown'

df_boolean = titanic[mask1 | mask2]
df_boolean.head()
```

```python
isin_filter = titanic['embark_town'].isin(['Southampton', 'Queenstown'])
df_isin = titanic[isin_filter]
df_isin.head()
```
