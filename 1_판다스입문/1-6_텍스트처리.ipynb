```python
# 문자열 메서드

import pandas as pd
import numpy as np

list_data = ["Apple_사과", "Banana_바나나", "Cherry_체리", np.nan]

ser = pd.Series(
    list_data,
    index = ['First ', ' Second ', ' Third', 'Fourth'],
    dtype = 'string'
)

print(ser)
print()

ser2 = pd.Series(
    list_data,
    index = ['First ', ' Second ', ' Third', 'Fourth'],
)

print(ser2)
```

```python
# 소문자로 변형 -> str접근자.lower()

print(ser.str.lower(), '\n')

# 대문자로 변형 -> str접근자.upper()

print(ser.str.upper(), "\n")

# 문자열 길이 계산, str접근자.len()
# len(str) => 내장함수 len(): 객체의 행의 수를 구할 때
print(ser.str.len())
```

```python
# 문자열 분할

exam1 = ser.str.split('_')
print(exam1)
type(exam1)
```

```python
# 문자열을 분할하여 데이터프레임으로 확장

exam2 = ser.str.split("_", expand = True)
print(exam2)
print(type(exam2))
```

```python
# 문자열 중 특정 요소 선택

ser.str.split("_").str.get(1)
```

```python
# 행 인덱스 추출

idx = ser.index
idx

print(idx.str.strip())
```

```python
# 문자열 교체("_"을 찾아서 ":"으로 변경)

ser.str.replace("_", ":", regex = False)
```

```python
ser.str.replace("[^a-zA-Z\\s]", "", regex = True)
```

```python
# 접근자 인덱싱 -> str[a]

ser.str[0]
```

```python
# 접근자 슬라이싱 -> str[a:b]

print(ser)
ser.str[0:4]
```

```python
# 문자열 메서드(contains)
# 특정 문자 포함 여부 확인, NA 값을 False로 처리

print(ser)
print()

contains_A = ser.str.contains("A", na = False)

print(contains_A)
print()

contains_pattern = ser.str.contains(r"[A|B][a-z]+")
print(contains_pattern)
print()

fullmatch_pattern = ser.str.fullmatch(r"[A|B][a-z]+")
print(fullmatch_pattern)
```

```python
# 새로운 시리즈 생성

list_data = ["a1", 'b2', 'c3', 'd4a5', 'e6e7', 'f8']
ser3 = pd.Series(list_data, dtype='string')

print(ser3)

digits = ser3.str.extractall(r"(\d)").groupby(level=0).agg(''.join)
print(digits)
print()

letters = ser3.str.extractall(r"([a-z])").groupby(level = 0).agg(''.join)
print(letters)
print()

split_columns = ser3.str.extract(r"(?P<Letter>[a-z]+)(?P<Ditgit>\d+)")
print(split_columns)
```
