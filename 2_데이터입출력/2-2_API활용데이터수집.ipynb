```python
# 

import requests
import pandas as pd

my_key = ""

def get_geocoding(place):
    url = f"https://dapi.kakao.com/v2/local/search/address.json?query={place}"
    headers = {"Authorization": f"KakaoAK {my_key}"}
    response = requests.get(url, headers = headers)
    result = response.json()

    # print(result["documents"][0]["y"])
    # print(result["documents"][0]["x"])
    return result["documents"][0]["y"], result["documents"][0]["x"]

lat = []
lng = []

places = ["대구 북구 대학로 80",
          "경북 구미시 대학로 61",
          "경북대학교",
          "대구 남구 명덕로 104"]

i = 0

for place in places:
    i += 1
    try:
        print(i, place)
        place_lat, place_lon = get_geocoding(place)
        lat.append(place_lat)
        lng.append(place_lon)

    except:
        lat.append('')
        lng.append('')
        print(i)

dict_data = {
    '위도': lat,
    '경도': lng
}

df = pd.DataFrame(dict_data, index=places)
df 
```

```python
# data.go.kr
# 국토교통부 아파트 매매 실거래가 자료

import requests
import xml.etree.ElementTree as ET
import pandas as pd

SERVICE_KEY = ""
url = "https://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade?"

def fetch_apt_trade(servicekey, LAWD_CD, DEAL_YMD):
    params = {
        'serviceKey' : servicekey,
        'LAWD_CD' : LAWD_CD,
        'DEAL_YMD': DEAL_YMD
    }
    
    response = requests.get(url, params = params)
    # print(response.content)
    root = ET.fromstring(response.content)
    # print(root)
    items = root.findall(".//item")
    # print(items)
    records = []
    for item in items:
        row = {}

        for child in item:
            row[child.tag] = child.text

        records.append(row)

    return records


data = fetch_apt_trade(SERVICE_KEY, LAWD_CD = "27200", DEAL_YMD = '202604')

df = pd.DataFrame(data)
print(df.shape)
df
```
