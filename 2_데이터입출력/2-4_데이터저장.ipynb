```python
# CSV 파일로 저장 
## df객체.to_csv("파일 이름(경로)")

import pandas as pd


data = {'name': ['Jerry', 'Riah', 'Paul'],
        'algol': ['A', 'A+', 'B'],
        'basic': ['C', 'B', 'B+'],
        'c++': ['B+', 'C', 'C+']
        }

df = pd.DataFrame(data)
df = df.set_index('name')
df

df.to_csv("./data/df_sample2.csv")
```

```python
df.to_csv("./data/df_sample_no_index2.csv", index=False)
```
