```python
# 데이터 개수 확인

import pandas as pd

df = pd.read_csv(r"C:/Users/KDT34/Documents/15-data-analyze/3_데이터살펴보기/data/auto-mpg.csv", header = None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']

df
```

```python
print(df.count())
print(type(df.count()))
```

```python
# 각 열 고유값 개수

# df객체['열 이름'].value_count()

df.head()
```

```python
unique_values = df['origin'].value_counts()
unique_values
```

```python
# 고유값의 상대적 비율 확인

unique_values_ratio = df['origin'].value_counts(normalize=True)
unique_values_ratio
```

```python
unique_values_percentage = (df['origin'].value_counts(normalize=True) * 100).round(1)
unique_values_percentage
```
