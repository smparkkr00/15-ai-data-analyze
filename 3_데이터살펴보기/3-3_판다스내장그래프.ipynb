```python
# 막대그래프
## df객체.plot(kind='bar)

import pandas as pd

df = pd.read_excel("./data/남북한발전전력량.xlsx")

df_ns = df.iloc[[0, 5], 3: ]

df_ns.index = ["South", "North"]
df_ns = df_ns.astype(int)
tdf_ns = df_ns.T
```

```python
tdf_ns.plot(kind = "bar")
```

```python
tdf_ns.plot(kind = "barh")
```

```python
tdf_ns.plot(kind = "bar", stacked=True)
```

```python
# 히스토그램
## df객체.plot(kind="hist")

tdf_ns.plot(kind = "hist")
```

```python
tdf_ns.plot(kind = "hist",
            alpha = 0.5,
            figsize= (8,4),
            grid= True,
            color='white',
            edgecolor='black')
```

```python
tdf_ns.hist(figsize=(6,2));
```

```python
tdf_ns.hist(bins = 20, grid=False, figsize=(6,2));
```
