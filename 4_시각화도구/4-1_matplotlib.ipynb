```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = "./data/malgun.ttf"

font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family = font_name)

df = pd.read_excel("./data/시도별_전출입_인구수.xlsx")
df = df.ffill()

mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')

df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis = 1)
df_seoul = df_seoul.rename({'전입지별': '전입지'}, axis = 1)
df_seoul = df_seoul.set_index('전입지')

sr_one = df_seoul.loc['경기도']
plt.plot(sr_one.index, sr_one.values)
plt.title('서울 -> 경기 인구이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

```python
sr_one = df_seoul.loc['경기도']

plt.figure(figsize=(14, 5))
plt.xticks(rotation="vertical")
plt.plot(sr_one.index, sr_one.values)

plt.title('서울 -> 경기 인구이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.legend(labels =['서울 -> 경기'], loc='best')

plt.show()
```

```python
sr_one = df_seoul.loc['경기도']
plt.style.use("ggplot")

plt.figure(figsize=(14,5))

plt.xticks(size=10, rotation="vertical")

plt.plot(sr_one.index, sr_one.values,
        marker = 'D',
        markerfacecolor = "red",
        markeredgecolor = "blue",
        markeredgewidth = 2,
        markersize = 10)

plt.title('서울 -> 경기 인구이동', size=30)
plt.xlabel('기간', size=20)
plt.ylabel('이동 인구수', size=20)

plt.legend(labels = ['서울 -> 경기'], loc='best', fontsize = 15)

plt.show()
```

```python
plt.rcParams["axes.unicode_minus"]=False

df = pd.read_excel("./data/남북한발전전력량.xlsx")
df.head()
df = df.loc[5:9]

df.drop('전력량 (억㎾h)', axis='columns', inplace =True)
df.set_index('발전 전력별', inplace=True)
```

```python
df= df.T
df
```

```python
for col in df.columns:
    df[col] = df[col].replace('-','0')


df = df.astype(float)

df = df.rename(columns={'합계':'총발전량'})
df
df['총발전량 - 1년'] = df['총발전량'].shift(1)
df['증감율'] = ((df['총발전량'] / df['총발전량 - 1년']) - 1)* 100

ax1 = df[['수력','화력']].plot(kind='bar', figsize=(20, 10 ), width=0.7, stacked=True)
ax2 = ax1.twinx()
ax2.plot(df.index, df.증감율, ls='--', marker='o', markersize=20, color='green', label='전년대비 증감율(%)')

ax1.set_ylim(0,500)
ax2.set_ylim(-50, 50)

plt.show()
```

```python
# 산점도

import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('default')

df = pd.read_csv('./data/auto-mpg.csv', header=None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']

df.plot(kind='scatter', x='weight', y='mpg', c='coral', s=10, figsize=(10,5))
plt.title('Scatter plot - mpg vs weight')
plt.show()
```

```python
cylinders_size = (df.cylinders/df.cylinders.max()) * 300

df.plot(kind='scatter', x='weight', y='mpg', c='coral', figsize=(10,5), s = cylinders_size,alpha = 0.3)
plt.title('Scatter plot - mpg vs weight')
plt.show()
```

```python
cylinders_size = (df['cylinders']/df['cylinders'].max()) * 300

df.plot(kind='scatter', x='weight', y='mpg', marker='+', figsize=(10,5), cmap= 'viridis', c = cylinders_size, s= 50, alpha = 0.7)

plt.savefig('./data/scatter.png')
plt.savefig('./data/scatter_transparent.png', transparent=True)
plt.title('Scatter plot - mpg vs weight')
plt.show()
```

```python
# 파이 차트

import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('default')

df = pd.read_csv('./data/auto-mpg.csv', header=None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']

df['count'] = 1
df.head(10)
```

```python
df_origin = df.groupby('origin').sum(numeric_only=True)
df_origin.index = ['USA', "EU", 'JAPAN']
df_origin
```

```python
df_origin['count'].plot(kind='pie', figsize=(7,5), autopct='%1.1f%%', startangle=27, colors=['red', 'blue', 'yellow'])

plt.title('Model Origin', size=20)
plt.axis('equal')
plt.legend(labels=df_origin.index, loc="upper right")
plt.show()
```

```python
# 박스 플롯(이상치 판별)

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

plt.style.use('grayscale')

df = pd.read_csv('./data/auto-mpg.csv', header=None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']

plt.rcParams['axes.unicode_minus']=False

font_path = './data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax1.boxplot(x=[df[df['origin']==1]['mpg'],
               df[df['origin']==2]['mpg'],
               df[df['origin']==3]['mpg']],
               labels = ['USA', 'EU', 'JAPAN'])

ax2.boxplot(x=[df[df['origin']==1]['mpg'],
               df[df['origin']==2]['mpg'],
               df[df['origin']==3]['mpg']],
               labels = ['USA', 'EU', 'JAPAN'],
               vert=False)

ax1.set_title('제조국가별 연비 분포(수직 박스 플롯)')
ax2.set_title('제조국가별 연비 분포(수평 박스 플롯)')

plt.show()
```
