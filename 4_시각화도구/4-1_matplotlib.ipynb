```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = "./data/malgun.ttf"

font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family = font_name)

df = pd.read_excel("./data/시도별_전출입_인구수.xlsx")
df = df.ffill()

mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')

df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis = 1)
df_seoul = df_seoul.rename({'전입지별': '전입지'}, axis = 1)
df_seoul = df_seoul.set_index('전입지')

sr_one = df_seoul.loc['경기도']
plt.plot(sr_one.index, sr_one.values)
plt.title('서울 -> 경기 인구이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

```python
sr_one = df_seoul.loc['경기도']

plt.figure(figsize=(14, 5))
plt.xticks(rotation="vertical")
plt.plot(sr_one.index, sr_one.values)

plt.title('서울 -> 경기 인구이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.legend(labels =['서울 -> 경기'], loc='best')

plt.show()
```

```python
sr_one = df_seoul.loc['경기도']
plt.style.use("ggplot")

plt.figure(figsize=(14,5))

plt.xticks(size=10, rotation="vertical")

plt.plot(sr_one.index, sr_one.values,
        marker = 'D',
        markerfacecolor = "red",
        markeredgecolor = "blue",
        markeredgewidth = 2,
        markersize = 10)

plt.title('서울 -> 경기 인구이동', size=30)
plt.xlabel('기간', size=20)
plt.ylabel('이동 인구수', size=20)

plt.legend(labels = ['서울 -> 경기'], loc='best', fontsize = 15)

plt.show()
```

```python
# 스타일리스트 출력

plt.style.use("ggplot")

plt.figure(figsize=(14,5))

plt.plot(sr_one.index, sr_one.values, marker = 'D', markersize = 10)

plt.title('서울 -> 경기 인구이동', size=30)
plt.xlabel('기간', size=20)
plt.ylabel('이동 인구수', size=20)

plt.legend(labels = ['서울 -> 경기'], loc='best', fontsize = 15)

plt.ylim(50000, 800000)

plt.annotate('',
            xy=(20, 620000),
            xytext=(2,290000),
            xycoords='data',
            arrowprops=dict(arrowstyle="->", color='skyblue', lw=5))

plt.annotate('',
            xy=(47, 450000),
            xytext=(30,580000),
            xycoords='data',
            arrowprops=dict(arrowstyle="->", color='olive', lw=5))

plt.annotate('인구이동 증가(1970-1995)',
            xy = (10,350000),
            rotation = 25,
            va='baseline',
            ha='center',
            fontsize=15)

plt.annotate('인구이동 감소(1995-2017)',
            xy = (40,470000),
            rotation = -11,
            va='baseline',
            ha='center',
            fontsize=15)
# plt.show()
```

```python
# Figure 객체 vs Axes 객체

import pandas as pd
import matplotlib.pyplot as plt

samsaung_revenue = pd.read_csv("./data/삼성전자_분기별_매출액.csv")
samsaung_revenue = samsaung_revenue.sort_values('quarter')
samsaung_revenue
```

```python
fig, axe = plt.subplots()
print(fig)
print(axe)
```

```python
fig, axe = plt.subplots(figsize=(8,2))
axe.plot(samsaung_revenue['quarter'], samsaung_revenue['value'])
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,2))

axes[0].plot(samsaung_revenue['quarter'], samsaung_revenue['value'])
samsaung_revenue['value'].plot(ax=axes[1])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,2), constrained_layout = True)

axes[0,1].plot(samsaung_revenue['quarter'], samsaung_revenue['value'])
plt.show()
```

```python
fig, axes = plt.subplot_mosaic([['top_left', 'right'],
                                ['bottom_left', 'right']],
                                figsize = (12,4))

axes['right'].plot(samsaung_revenue['quarter'], samsaung_revenue['value'])
plt.show()
```
