```python
import pandas as pd

df = pd.DataFrame({
    'c1':['a', 'a', 'b', 'a', 'b'],
    'c2': [1, 1, 1, 2, 2],
    'c3': [1, 1, 2, 2, 2]
})

df
```

```python
# duplicated() 매서드

df_dup_first = df.duplicated()
df_dup_first
```

```python
df_dup_last = df.duplicated(keep='last')
df_dup_last
```

```python
df["c2"].duplicated()
```

```python
# 중복 데이터 제거
print(df)
print()
df2 = df.drop_duplicates()
print(df2)
```

```python
print(df)
print()
df2 = df.drop_duplicates(keep = 'last')
print(df2)
```

```python
print(df)
df4 = df.drop_duplicates(keep=False)

print()
print(df4)
```

```python
print(df)
df5 = df.drop_duplicates(subset=['c2', 'c3'])
print()
print(df5)
```

```python
print(df)
df6 = df.drop_duplicates(subset=['c2', 'c3'], keep=False)
print()
print(df6)
```
