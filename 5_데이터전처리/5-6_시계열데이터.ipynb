```python
# period

import pandas as pd

dates = ['2019-01-01', '2020-03-01', '2020-06-01']

ts_dates = pd.to_datetime(dates)

ts_dates
```

```python
pr_day = ts_dates.to_period(freq="D")
pr_month = ts_dates.to_period(freq="M")
pr_year = ts_dates.to_period(freq="Y")

print(pr_day)
print(pr_month)
print(pr_year)
```

```python
# 시계열 데이터 만들기
ts_ms = pd.date_range(
    start="2024-01-01", 
    end=None, 
    periods=6, 
    freq="MS",
    tz='Asia/Seoul'
)
print(ts_ms)
```
