```python
# period

import pandas as pd

dates = ['2019-01-01', '2020-03-01', '2021-06-01']

ts_dates = pd.to_datetime(dates)
print(ts_dates)
```

```python
pr_day=ts_dates.to_period(freq='D')
print(pr_day)
pr_month=ts_dates.to_period(freq='M')
print(pr_month)
pr_year=ts_dates.to_period(freq='Y')
print(pr_year)
```

```python
# 시계열 데이터 만들기

ts_ms = pd.date_range(
                start = '2024-01-01',
                end=None,
                periods=6,
                freq='MS',
                tz='Asia/Seoul'
                )

print(ts_ms)
```

```python
ts_2ms = pd.date_range(
                start = '2024-01-01',
                end=None,
                periods=6,
                freq='ME',
                tz='Asia/Seoul'
                )

print(ts_2ms)
```

```python
ts_3ms = pd.date_range(
                start = '2024-01-01',
                end=None,
                periods=6,
                freq='3MS',
                tz='Asia/Seoul'
                )

print(ts_3ms)
```

```python
pr_m = ts_ms = pd.period_range(
                start = '2024-01-01',
                end=None,
                periods=3,
                freq='M',
                )

print(pr_m)
```
