```python
# 문제1

import pandas as pd

score_list = [85, 92, 78, 90, 88]
name_list = ['Alice', 'Bob', 'Charlie', "David", 'Eve']

score = pd.Series(score_list)

print(score, '\n')

score_name = pd.Series(score_list, index = name_list)

print(score_name, '\n')

##############################################################

sub_score_dict = {
    'Python': 95,
    'SQL': 85,
    'Web': 90
}

sub_score = pd.Series(sub_score_dict)
print(sub_score)
```
